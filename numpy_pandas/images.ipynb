```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_logo():
    image = plt.imread('hermes.png')
    logo = image[: , : , 0]
    logo[logo == 1] = 255
    return logo

logo = load_logo()
print("Shape of image is " , logo.shape)
print(logo)
```

```python
def plot_image(img):
    plt.imshow(img , cmap='binary')

plot_image(logo)
```

```python
cropped_logo = logo[: , :140]
plot_image(cropped_logo)
```

```python
cropped_logo = logo[: , 140:]
plot_image(cropped_logo)
```

```python
cropped_logo_R = cropped_logo.copy()
cropped_logo_G = cropped_logo.copy()
cropped_logo_B = cropped_logo.copy()
```

```python
plot_image(cropped_logo[14:26 , 14:25])
```

```python
row_start = 14
row_end = 26
col_start = 14
col_end = 25
```

```python
cropped_logo_R[row_start:row_end , col_start : col_end] = 25
cropped_logo_G[row_start:row_end , col_start : col_end] = 15
cropped_logo_B[row_start:row_end , col_start : col_end] = 0
```

```python
img_final = (np.dstack([cropped_logo_R,cropped_logo_G,cropped_logo_B]))
```

```python
plt.imshow(img_final)
```
